```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'/Users/linyu/Desktop/Lin Yu Python Pandas /Ice Cream Ratings.csv')
df.set_index('Date')
```

```python
# we use plot to visualize things in pandas
df.plot()# Line is the default
```

```python
print(plt.style.available)
```

```python
plt.style.use('fivethirtyeight') # you can change the styles
```

```python
df.plot(kind='line', title='ice cream rating')
```

```python
df.plot(kind='line',subplots=True) #break the y into 3 different plots
```

```python
df.plot.barh(stacked=True) #change to horizontal charts
```
